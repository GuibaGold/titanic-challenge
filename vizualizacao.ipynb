#Importando bibliotecas para leitura dos dados e para vizualização dos dados

```python
import pandas as panda
import matplotlib.pyplot as plot
```

Lendo os dados e criando a Figura para vizualização

```python
quadro = panda.read_csv("train.csv")
figure = plot.figure(figsize=(20,8))
```

```python
#Porcentagem de passageiros que sobreviveram ou não em gráfico de barra
plot.subplot2grid((2,3), (0,0))
quadro.Survived.value_counts(normalize=True).plot(kind="bar", alpha=0.5)
plot.title("Sobreviveu")
```

```python
#Scatter da idade dos passageiros que sobreviveram ou não
plot.subplot2grid((2,3), (0,1))
plot.scatter(quadro.Survived, quadro.Age, alpha=0.1)
plot.title("Idade que Sobreviveu")
```

```python
#Porcentagem dos passageiros na Primeira, Segunda e Terceira classe do navio
plot.subplot2grid((2,3), (0,2))
quadro.Pclass.value_counts(normalize=True).plot(kind="bar", alpha=0.5)
plot.title("Classe Navio")
```

```python
#Estimativa de densidade Kernel da idade dos passageiros que sobreviveram e que morreram
plot.subplot2grid((2,3), (1,0), colspan=2)
for x in [1,2,3]:
    quadro.Age[quadro.Pclass == x].plot(kind="kde")

plot.title("Classe e Idade dos passageiros")
plot.legend(("Primeira classe", "Segunda classe", "Terceira classe"))
```

```python
#Local de Embarque dos passageiros
plot.subplot2grid((2,3), (1,2))
quadro.Embarked[quadro.Survived == 1].value_counts(normalize=True).plot(kind="bar", alpha=0.5)
plot.title("Porcentagem de passageiros que embacaram em 'Southempton', 'Queenstown' e 'Cherboug' e sobreviveram")
```

```python
plot.subplot2grid((2,3), (1,2))
quadro.Embarked.value_counts(normalize=True).plot(kind="bar", alpha=0.5)
plot.title("Porcentagem de passageiros por local de embarque")
```

Pode-se observar que o local de embarque e a idade não influenciou de maneira significativa na sobrevivência dos passageiros, porém a classe em que os passageiros estavam pode teve sim relevância.


Agora vamos analizar pelo gênero de cada um dos passageiros:

```python
import pandas as panda
import matplotlib.pyplot as plot

female_color = "#FA0000"
```

```python
quadro = panda.read_csv("train.csv")
figure = plot.figure(figsize=(20,8))
```

```python
#Porcentagem de passageiros que sobreviveram ou não em gráfico de barra
plot.subplot2grid((3,4), (0,0))
quadro.Survived.value_counts(normalize=True).plot(kind="bar", alpha=0.5)
plot.title("Sobreviveu")
```

```python
plot.subplot2grid((3,4), (0,1))
quadro.Survived[quadro.Sex == "male"].value_counts(normalize=True).plot(kind="bar", alpha=0.5)
plot.title("Homens que Sobreviveram")

plot.subplot2grid((3,4), (0,3))
quadro.Survived[quadro.Sex == "female"].value_counts(normalize=True).plot(kind="bar", alpha=0.5, color = female_color)
plot.title("Mulheres que Sobreviveram")
```

```python

plot.subplot2grid((3,4), (1,0))
quadro.Sex[quadro.Survived == 1].value_counts(normalize=True).plot(kind="bar", alpha=0.5, color=[female_color, 'b'])
plot.title("Porcentagem de Homens e Mulheres que Sobreviveram")
```

Agora está claro um fator contribuinte para sobrevivência de um Passageiro: Ser do genero Feminino

Vamos analizar a sobrevivência em relação a classe de passageiro:

```python
plot.subplot2grid((3,4), (1,1), colspan=4)
for x in [1,2,3]:
    quadro.Survived[quadro.Pclass == x].plot(kind="kde")

plot.title("Estimativa de Densidade Classe")
plot.legend(("Primeira classe", "Segunda classe", "Terceira classe"))
```

Pode-se notar que a maior taxa de mortalidade está associada a terceira classe de passageiros do Titanic.

Agora sabendo que a classe onde o passageiro estava localizado realmente teve influência sobre a sobrevivência, podemos fazer uma comparação entre Homens ricos "Primeira Classe" e Homens pobres "Terceira Classe":

```python
plot.subplot2grid((3,4), (2,0))
quadro.Survived[(quadro.Sex == "male") & (quadro.Pclass == 1)].value_counts(normalize=True).plot(kind="bar", alpha=0.5)
plot.title("Homens Ricos que Sobreviveram -")
```

```python
plot.subplot2grid((3,4), (2,1))
quadro.Survived[(quadro.Sex == "male") & (quadro.Pclass == 3)].value_counts(normalize=True).plot(kind="bar", alpha=0.5)
plot.title("Homens Pobres que Sobreviveram -")
```

Faremos o mesmo para Mulheres, comparando-as entre as Classes que estavam inseridas:


```python
plot.subplot2grid((3,4), (2,0))
quadro.Survived[(quadro.Sex == "female") & (quadro.Pclass == 1)].value_counts(normalize=True).plot(kind="bar", alpha=0.5, color=female_color)
plot.title("Mulheres Ricas que Sobreviveram -")

plot.subplot2grid((3,4), (2,3))
quadro.Survived[(quadro.Sex == "female") & (quadro.Pclass == 3)].value_counts(normalize=True).plot(kind="bar", alpha=0.5, color= female_color)
plot.title("Mulheres Pobres que Sobreviveram -")
```

Percebemos então o primeiro fator fundamental na sobrevivência de um passageiro: Ser Mulher e estar na Primeira Classe
